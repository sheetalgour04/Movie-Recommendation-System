```python
def helper(obj):
    
    keyw = []
    for i in obj:
        
        keyw.append(i['name'])
        
    return keyw
```

```python
tags = helper(ast.literal_eval(movies.iloc[0]['genres']))
```

```python
tags
# movies['genres']
```

```python
def convert(obj):
    
    keyw = []
    for i in ast.literal_eval(obj):
        
        keyw.append(i['name'])
        
    return keyw
    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies
```

```python
movies.iloc[0]['keywords']
```

```python
movies['keywords'] = movies['keywords'].apply(convert) 
```

```python
movies
```

```python
ast.literal_eval(movies.iloc[0]['cast'])
```

```python
def convert2(obj):
    c = 0
    keyw = []
    for i in ast.literal_eval(obj):
        
        if c != 3:
            keyw.append(i['name'])
            c += 1
        
    return keyw
    
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies
```

```python
ast.literal_eval(movies.iloc[0]['crew'])
```

```python
def convert3(obj):
    
    keyw = []
    for i in ast.literal_eval(obj):
        
        if i['job'] == 'Director':
            keyw.append(i['name'])
        
    return keyw
    
```

```python
movies['crew'] = movies['crew'].apply(convert3)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(' ','') for i in x] )
movies['cast']=movies['cast'].apply(lambda x: [i.replace(' ','') for i in x] )
movies['crew']=movies['crew'].apply(lambda x: [i.replace(' ','') for i in x] )
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(' ','') for i in x] )
```

```python
movies.head(1)
```

```python

movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew'] + movies['keywords']
```

```python
movies.drop(['overview','keywords','cast','crew','genres'] , axis=1,inplace=True)
```

```python
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(x))
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

# Vectorization

```python
movies.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english' , max_features= 5000)
```

```python
vector= cv.fit_transform(movies['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
import nltk

from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('marine')
```

```python
def stem(s):
    
    l = []
    for i in s.split():
        l.append(ps.stem(i))
        
    return ' '.join(l)
    
```

```python
stem("in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction samworthington zoesaldana sigourneyweaver jamescameron cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d")
```

```python
movies["tags"] = movies['tags'].apply(lambda x : stem(x))
```

```python
vector = cv.fit_transform(movies['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie_name):
    
#     if movie_name not in movies['title']:
#         print("Ooops !!! Movie data not found\n")
#         return 
    
    movie_index = movies[movies['title'] == movie_name].index[0]
    distances = similarity[movie_index]
    
    top_5 = sorted(list(enumerate(distances)),reverse=True , key=lambda x: x[1])[1:6]
    
    for i in top_5:
        print(movies.title.iloc[i[0]])
```
